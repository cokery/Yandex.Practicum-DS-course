```python
def replace_na_area(column_name):
    display('N/A percentage before replacement in '+column_name, df[column_name].isna().mean() * 100)
    df[column_name] = df[column_name+'_temp']
    df_temp = df[~df[column_name].isna()]
    rate = np.mean(df_temp[column_name]/df_temp['total_area'])

    def fill_kitchen_area(row):
        if pd.isna(row[column_name]):
            return rate * row['total_area']
        else:
            return row[column_name]

    df[column_name] = df.apply(fill_kitchen_area, axis=1)
    display('N/A percentage after replacement in '+column_name, df[column_name].isna().mean()  * 100)

replace_na_area('kitchen_area')
replace_na_area('living_area')
df.drop(columns=['kitchen_area_temp', 'living_area_temp'])

df.query('total_area * 0.9 > (living_area + kitchen_area)')
```

## Расчёты и добавление результатов в таблицу

```python
try:
    df['price_per_sqmeter'] = df['last_price'] /df['total_area'] 
except:
    print('Где-то есть ноль в общей площади')
    
try:
    df['kitchen_area_per_total'] = df['kitchen_area'] /df['total_area'] 
except:
    print('Где-то есть ноль в общей площади')

df['day'] = (df['first_day_exposition'].dt.weekday
            .map({0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thirsday', 4:'Friday', 5:'Saturday', 6:'Sunday'})
            )
df['month'] = df['first_day_exposition'].dt.month
df['year'] = df['first_day_exposition'].dt.year

   
def floor_group(row):
    if row['floor'] == 1:
        return 'первый'
    elif row['floor'] == row.loc['floors_total']:
        return 'последний'
    else:
        return 'другой'

df['floor_group'] = df.apply(floor_group, axis=1)
display(df.query('floor_group == "последний"').head())
display(df.query('floor_group == "первый"').head())
display(df.query('floor_group == "другой"').head())
```

2. Зависимость от числа комнат: чем больше комнат, тем выше стоимость. 
3. Зависимость от удалённости от центра: в общем объеме данных сильной корреляции с ценой нет, т.к. присутствуют отдельные небольшие населенные пункты, где цена значительно ниже, чем в крупных при той же или меньше удаленности от цента.
